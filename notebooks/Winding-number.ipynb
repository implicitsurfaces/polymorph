```python
import numpy as np

def winding_number_indefinite_integral(t, R, x, y):
    R2 = R * R
    sint = np.sin(t/2)
    cost = np.cos(t/2)

    term1 = R2 * sint + 2 * R * (x * sint - y * cost)
    term2 = (x**2 + y**2) * sint
    
    numerator = 1 / cost * (term1 + term2)
    denominator = R2 - x**2 - y**2
    
    arctan_term = np.arctan(numerator / denominator)
    
    return (R * (arctan_term / R + t / (2 * R))) / (2 * np.pi)
    
# Example usage:
R = 1.2


def w_n(x, y):
    return winding_number_indefinite_integral(np.pi, R, x, y) -  winding_number_indefinite_integral(np.pi / 4 + 0.1, R, x, y)
    
```

```python
import matplotlib.pyplot as plt
def render(bounds=(-3, 3), n=500):
    x = np.linspace(bounds[0], bounds[1], n)
    X, Y = np.meshgrid(x, x)

    grid_points = np.column_stack((X.flatten(), Y.flatten()))
    plt.imshow(w_n(X, Y).reshape(n, n), cmap="coolwarm", origin="lower", extent=[bounds[0], bounds[1],bounds[0], bounds[1]])
```
